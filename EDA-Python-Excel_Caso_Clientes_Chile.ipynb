# **CREAR CAMPO CALCULADO CONDICIONAL:**

Se requiere conocer si las tiendas son RETAIL O E-TAILERS, para ello tener en cuenta la condición de negocio: si es BIP, HPONLINE, MERCADOLIBRE, SPDIGITAL y WEI son categoría “E-TAILERS” el resto son RETAIL.

Solución: Crear un Campo Calculado "Tipo_Tienda" a partir del valor en "TiendaPm"

image.png

```python
def determinar_tipo_tienda(tienda):
    if tienda in ["BIP", "HPONLINE", "MERCADOLIBRE", "SPDIGITAL", "WEI"]:
        return "E-TAILERS"
    else:
        return "RETAIL"

consolidado_df['Tipo_Tienda'] = consolidado_df['TiendaPm'].apply(determinar_tipo_tienda)
```
